```python
# define half moon data

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=300, noise=0.1, random_state=0)

# define the figure
fig = plt.figure()
plt.scatter(X[:,0], X[:,1], c=y)
# plot the data
plt.scatter(X[:,0], X[:,1], c=y)
# make the set of the abscisses of X that have y=1
X0 = X[y==1]
# make the set of the abscisses of X that have y=0
X1 = X[y==0]

haut_x = X1[:,0]
haut_x2 = haut_x ** 2
haut_x3 = haut_x ** 3
haut_y = X1[:,1]
haut_b = np.ones(len(haut_x))

bas_x = X0[:,0]
bas_x2 = bas_x ** 2
bas_x3 = bas_x ** 3
bas_y = X0[:,1]
bas_b = np.ones(len(bas_x))
print(haut_x)


plt.show()
```

```python
from sklearn.model_selection import train_test_split
import random

haut = list(zip(haut_x, haut_x2, haut_x3, haut_y, haut_b))
bas = list(zip(bas_x, bas_x2, bas_x3, bas_y, bas_b))

# labelling oranges with 0 and lemons with 1:
labelled_data = list(zip(haut + bas, 
                         [0] * len(haut) + [1] * len(bas)))
random.shuffle(labelled_data)

data, labels = zip(*labelled_data)

res = train_test_split(data, labels, 
                       train_size=0.8,
                       test_size=0.2,
                       random_state=42)
train_data, test_data, train_labels, test_labels = res    
print(train_data[:10], train_labels[:10])
```

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def d_sig(x):
    return sigmoid(x) * (1 - sigmoid(x))

class Perceptron:
    def __init__(self, weights, learning_rate=0.1):
        self.weights = np.array(weights)
        self.learning_rate = learning_rate

    @staticmethod
    def unit_step_function(x):
        if  x < 0:
            return 0
        else:
            return 1
        
    def __call__(self, in_data):
        weighted_input = self.weights * in_data
        weighted_sum = weighted_input.sum()
        return Perceptron.unit_step_function(weighted_sum)

    # train the perceptron using gradient descent and loss function expected output - sigmoid ( y - a0 * x - a1 * x - a2 * x - a3)
    def loss_function(self, target, prediction, in_data):
        w1, w2, w3, w4, w5 = self.weights
        a0 = -w1/w4
        a1 = -w2/w4
        a2 = -w3/w4
        b = -w5/w4
        y = in_data[3]
        y_poly = a0 * in_data[0] ** 3 + a1 * in_data[0] ** 2 + a2 * in_data[0] + b
        #y_poly = w1 * in_data[0] ** 3 + w2 * in_data[0] ** 2 + w3 * in_data[0] + w5
        return -(sigmoid(y - y_poly)- target) ** 2


    def update_weights(self, target, prediction, in_data):
        loss = self.loss_function(target, prediction, in_data)
        w1, w2, w3, w4, w5 = self.weights
        a0 = -w1/w4
        a1 = -w2/w4
        a2 = -w3/w4
        b = -w5/w4
        y = in_data[3]
        y_poly = a0 * in_data[0] ** 3 + a1 * in_data[0] ** 2 + a2 * in_data[0] + b
        #y_poly = w1 * in_data[0] ** 3 + w2 * in_data[0] ** 2 + w3 * in_data[0] + w5
        # gradient of loss function with respect to weights
        dL_dw1 = -2 * loss *d_sig(y - y_poly) * (in_data[0] ** 3)
        dL_dw2 = -2 * loss *d_sig(y - y_poly) * (in_data[0] ** 2)
        dL_dw3 = -2 * loss *d_sig(y - y_poly) * (in_data[0])
        dL_dw5 = -2 * loss *d_sig(y - y_poly) * (1)
        dL_dw4 = -2 * loss *d_sig(y - y_poly) * (a0 * in_data[0] ** 3 + a1 * in_data[0] ** 2 + a2* in_data[0] + b - y)

        self.weights[0] -= self.learning_rate * dL_dw1
        self.weights[1] -= self.learning_rate * dL_dw2
        self.weights[2] -= self.learning_rate * dL_dw3
        self.weights[4] -= self.learning_rate * dL_dw5
        self.weights[3] -= self.learning_rate * dL_dw4
        
        Xp = np.linspace(-1.5, 2, 100)
        Yp = a0 * Xp ** 3 + a1 * Xp ** 2 + a2 * Xp + b  
        plt.plot(Xp, Yp, color='red')

                
    def evaluate(self, data, labels):
        evaluation = Counter()
        for index in range(len(data)):
            label = int(round(p(data[index]),0))
            if label == labels[index]:
                evaluation["correct"] += 1
            else:
                evaluation["wrong"] += 1
        return evaluation
                

p = Perceptron(weights = np.random.rand(5),
               learning_rate=0.01)

print("train" ,train_data)
print("label", train_labels)
for index in range(len(train_data)):
    p.update_weights(train_labels[index], 
             p(train_data[index]), 
             train_data[index])
    
evaluation = p.evaluate(train_data, train_labels)
print(evaluation.most_common())
evaluation = p.evaluate(test_data, test_labels)
print(evaluation.most_common())

print(p.weights)
# modify y axis limit
plt.ylim(-1.5, 1.5)
plt.scatter(haut_x, haut_y, color='orange')
plt.scatter(bas_x, bas_y, color='blue')

plt.show()


    
```
