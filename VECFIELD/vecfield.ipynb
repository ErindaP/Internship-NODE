```python
# Path: vecfield.ipynb
# Now we want to place 3 random points in the grid, the line between the first point and the two others will be the line that will divide the grid in 4 spaces. We will then attribute a different vector field to each space and plot the result

import numpy as np
import matplotlib.pyplot as plt

# Create grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)

# Create meshgrid
X, Y = np.meshgrid(x, y)

# Place 3 random points in the grid and store them in a list
points = []
for i in range(3):
    x = np.random.uniform(-2,2)
    y = np.random.uniform(-2,2)
    points.append([x,y])

# Plot the points

for i in range(3):
    plt.scatter(points[i][0],points[i][1], color = 'blue')

# Define vector field
U1 = 1
V1 = 1
U2 = -1
V2 = 1
U3 = 1
V3 = -1
U4 = -1
V4 = -1


# Plot the function that links the 3 points  (y = ax + b) 
# We need to find a and b
# a = (y2-y1)/(x2-x1)
# b = y1 - a*x1

a1 = (points[2][1]-points[0][1])/(points[2][0]-points[0][0])
b1 = points[0][1] - a1*points[0][0]

a2 = (points[1][1]-points[0][1])/(points[1][0]-points[0][0])
b2 = points[0][1] - a2*points[0][0]



# Define the lines
x1 = np.linspace(-2, 2, 20)
y1 = a1*x1 + b1
x2 = np.linspace(-2, 2, 20)
y2 = a2*x2 + b2









# Plot the lines
plt.plot(x1,y1, color = 'black')
plt.plot(x2,y2, color = 'black')


# Now add a random blue point in the grid and depending on the position of the point, plot the vector field that corresponds to that space next to the point
"""
#create a random point
x = np.random.uniform(-2,2)
y = np.random.uniform(-2,2)

#plot the point
plt.scatter(x,y, color = 'blue')

# attribute the good vector to the point and display it in pink starting from the point
if x<points[0][0] and y<points[0][1]:
    plt.quiver(x, y, U1, V1, color = 'green')
elif x>points[0][0] and y>points[0][1]:
    plt.quiver(x, y, U2, V2, color = 'green')
elif x<points[0][0] and y>points[0][1]:
    plt.quiver(x, y, U3, V3, color = 'green')
elif x>points[0][0] and y<points[0][1]:
    plt.quiver(x, y, U4, V4, color = 'green')

"""

# Now add a random blue point in the grid and depending on the position of the point, plot the vector field that corresponds to that space next to the point

#create 10 random points
for i in range(20):
    x = np.random.uniform(-2,2)
    y = np.random.uniform(-2,2)

    #plot the point
    plt.scatter(x,y, color = 'blue')
    if y < a2*x+b2 and y<a1*x + b1:
            plt.quiver(x, y, U1, V1, color = 'green')
    elif y > a2*x+b2 and y>a1*x + b1:
            plt.quiver(x, y, U2, V2, color = 'green')
    elif y < a2*x+b2 and y>a1*x + b1:
            plt.quiver(x, y, U3, V3, color = 'green')
    elif y > a2*x+b2 and y<a1*x + b1:
            plt.quiver(x, y, U4, V4, color = 'green')

    

#Now we want to build a tool to see the trajectory of a point in the grid depending on the vector field
# We will use the Euler method to do so

# Define the function that will give the vector field depending on the position of the point
def vecfield(x,y):
    if y < a2*x+b2 and y<a1*x + b1:
        return U1,V1
    elif y > a2*x+b2 and y>a1*x + b1:
        return U2,V2
    elif y < a2*x+b2 and y>a1*x + b1:
        return U3,V3
    elif y > a2*x+b2 and y<a1*x + b1:
        return U4,V4
    
# Define the function that will give the trajectory of the point
def trajectory(x,y):
    # Define the step size
    h = 0.05
    # Define the number of steps
    n = 100
    # Create empty lists to store the points
    X = []
    Y = []
    # Add the initial point to the list
    X.append(x)
    Y.append(y)
    # Create a loop to add the next points to the list
    for i in range(n):
        # Define the vector field at the point
        u,v = vecfield(x,y)
        # Define the next point
        x = x + h*u
        y = y + h*v
        # Add the next point to the list
        X.append(x)
        Y.append(y)
    # Return the lists
    return X,Y

# Define the initial point
x = np.random.uniform(-2,2)
y = np.random.uniform(-2,2)

# Plot the point
plt.scatter(x,y, color = 'blue')

# Plot the trajectory
X,Y = trajectory(x,y)
plt.plot(X,Y, color = 'red')


# Block the size of the plot
plt.axis([-2, 2, -2, 2])

# Show the plot

    

plt.show()


# Now we want to see the point moving frame by frame in the grid
# For this we will use the animation tool of matplotlib

# Import the animation tool
from matplotlib import animation
# Define the figure
fig = plt.figure()
# Define the axes
ax = plt.axes(xlim=(-2, 2), ylim=(-2, 2))
line, = ax.plot([], [], 'bo',lw=2)
# Create an empty line
def init():
    line.set_data([], [])
    return line,
#define the initial point
x = np.random.uniform(-2,2)
y = np.random.uniform(-2,2)
X,Y = trajectory(x,y)
print(X)
# Define the function that will create the animation
def animate(i):
    x = X[i]
    y = Y[i]
    line.set_data(x, y)
    return line,
 

# Create the animation
anim = animation.FuncAnimation(fig, animate, frames=len(X), interval=20, blit=True, init_func=init)

# save the animation
anim.save('animation.mp4', fps=30, extra_args=['-vcodec', 'libx264'])







plt.show()




```

```python
# Path: vecfield.ipynb
# Now we want to find the tangent vector field of the half moon dataset

# First we need to find the tangent vector of each point of the extremity of the half moon
# To do so we will use the function "np.gradient" to find the gradient of the half moon
# We will use the x and y coordinates of each point to find the gradient of the half moon
# We will then use the gradient to find the tangent vector of each point

# Define the gradient of the half moon
gradient = np.gradient(X)

# Define the tangent vector of each point
tangent = []
for i in range(((np.size(X[:,1]))//2)  ):
    tangent.append(np.array([gradient[0][i], gradient[1][i]]))

print(np.size(tangent))
# Define the tangent vector field
tangent = np.array(tangent)
print(tangent)

# Define the x and y coordinates of the tangent vector  field
x = X[:,0]
y = X[:,1]

print(len(X))
print(np.size(tangent[:,0]))
print(np.size(tangent[:,1]))
print(np.size(x))
print(np.size(y))



# Plot the tangent vector field
plt.quiver(x,y,tangent[:,0],tangent[:,1])

# Plot the half moon
plt.scatter(X[:,0], X[:,1], c=y)

# Show the plot
plt.show()


# Fuze the points that are close to each other into one point
# To do so we will use the function "np.unique" to find the unique points of the half moon
# We will then use the function "np.where" to find the index of the points that are close to each other
# We will then use the function "np.mean" to find the mean of the points that are close to each other
# We will then use the function "np.delete" to delete the points that are close to each other
# We will then use the function "np.append" to add the mean of the points that are close to each other

# Define the unique points of the half moon
unique = np.unique(X, axis=0)

# Define the index of the points that are close to each other
index = np.where(np.isclose(X, unique, atol=1).all(axis=1))[0]

# Define the mean of the points that are close to each other
mean = np.mean(X[index], axis=0)

# Delete the points that are close to each other
X = np.delete(X, index, axis=0)

# Add the mean of the points that are close to each other
X = np.append(X, [mean], axis=0)

# Plot the half moon
plt.scatter(X[:,0], X[:,1])

# Show the plot
plt.show()
    
```

```python
def trajectory3(x,y):
    # Define the step size
    h = 0.1
    # Define the number of steps
    n = 20
    # Create empty lists to store the points
    X = []
    Y = []
    # Add the initial point to the list
    X.append(x)
    Y.append(y)
    # Create a loop to add the next points to the list
    for i in range(n):
        # Define the vector field at the point
        #print( (vecfield2(x,y)))
        #print("x = ", x)
        #print("y = ", y)
        
        u,v = vecfield3(x,y)
        # Define the next point
        x = x + h*u
        y = y + h*v
        # Add the next point to the list
        X.append(x)
        Y.append(y)
    # Return the lists
    return X,Y
```

## Other approach  : Lagrangian interpolation 


We will want to construct a polynomial that separates the two half-moons by choosing the minimum number of points possible.

```python
# generate half moon data
X,y = make_moons(n_samples=100, noise=0.1)
# plot the data
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
# defining function that will generate the Lagrangian polynomial that passes through the points

def lagrange(X):
    # define the number of points
    n = len(X)
    # define the polynomial
    def L(x):
        # define the polynomial
        p = 0
        # create a loop to add the terms
        for i in range(n):
            # define the term
            term = X[i,1]
            # create a loop to add the factors
            for j in range(n):
                if j != i:
                    term = term*(x-X[j,0])/(X[i,0]-X[j,0])
            # add the term to the polynomial
            p = p + term
        # return the polynomial
        return p
    # return the function
    return L

# generate the Lagrangian polynomial
# generate 3 random points
X = np.random.rand(4,2)
#plot the points
plt.scatter(X[:,0], X[:,1])
# generate the Lagrangian polynomial
L = lagrange(X)
# plot the polynomial
x = np.linspace(0,1,100)
plt.plot(x,L(x))
plt.show()

```

```python
X,y = make_moons(n_samples=100, noise=0.1)
# plot the data
plt.scatter(X[:,0], X[:,1], c=y)

# find the best points to use to generate the Lagrangian polynomial
# define the number of points
n = 4
# define the number of points to test
m = 1000
# define the error
error = 1000
# create a loop to find the best points
for i in range(m):
    # generate the points
    X0 = np.random.rand(n,2)
    # generate the Lagrangian polynomial
    L = lagrange(X0)
    # calculate the error
    error0 = np.mean(np.abs(L(X[:,0])-X[:,1]))
    # check if the error is smaller than the previous error
    if error0 < error:
        # update the error
        error = error0
        # update the points
        X1 = X0
# plot the points
plt.scatter(X1[:,0], X1[:,1], color = 'red')


# generate the Lagrangian polynomial
L = lagrange(X1)
# plot the polynomial
x = np.linspace(-1.5,2.5,100)
#set y axis limits
plt.ylim(-1,2)
plt.plot(x,L(x), color = 'red')
plt.show()




```
